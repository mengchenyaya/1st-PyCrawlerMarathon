* Hint: 將所有資料存在一個變數當中



```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.support.ui import Select,WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
#extension
options = webdriver.ChromeOptions()
options.add_extension('/Users/mengchen/ChroPath.crx')

#open
browser = webdriver.Chrome(executable_path='/Users/mengchen/chromedriver',options = options)
browser.get("http://taqm.epa.gov.tw/taqm/tw/MonthlyAverage.aspx")
browser.maximize_window()

#year
selectYear = Select(browser.find_element_by_id("ctl05_ddlYear"))
selectYear.select_by_value('2019')

#area
selectSite = Select(browser.find_element_by_id("ctl05_ddlSite"))
selectSite.select_by_value('58')

#search
browser.find_element_by_id('ctl05_btnQuery').click()
```

```python
#get table
WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'TABLE_G')))
result = browser.find_element_by_class_name('ResultPanel')
df = pd.read_html(result.get_attribute('innerHTML'))[0]

#parse
df = df.loc[:,['監測日期','監測項目','監測值']]
df = df.pivot(index='監測日期',columns='監測項目')
df.columns = df.columns.get_level_values(1)
df.index.rename('DATE',inplace=True)
df
```
