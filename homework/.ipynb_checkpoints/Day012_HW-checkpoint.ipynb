* 能夠利用 Request + BeatifulSoup 撰寫爬蟲，並存放到合適的資料結構


## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.html'
r = requests.get(url)

soup = BeautifulSoup(r.text)
raw_news = soup.find(class_='part_list_2').find_all('h3')
```

```python
raw_news
```

### ① 取出今天所有的發文

```python
for news in raw_news:
    print(news.find('span').text+' '+news.find('a').text+'\n')
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？
##### by pandas dataframe

```python
df = [[news.find('span').text,news.find('em').text,news.find('a').text.replace('\u3000',' ')] for news in raw_news]
df = pd.DataFrame(df,columns=['time','category','title'])
df
```

### ③ 哪一個類別的文章最多

```python
categories = df.groupby('category').size()
categories
```

```python
f'{categories.idxmax()} {categories.max()}'
```
