```python
import requests
import json
```

```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›žå‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

url = 'https://www.dcard.tw/_api/forums/pet/posts?popular=true'
res = requests.get(url)
url2 = 'https://www.dcard.tw/_api/forums/pet/posts?popular=false'
res2 = requests.get(url2)
jf = json.loads(res.text)
jf2 = json.loads(res2.text)
print(f'å›žå‚³ç­†æ•¸ï¼š{len(jf)}') 
print('åŒ…å«æ¬„ä½ï¼š')
for i in jf[0].keys() :
    print(i)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
all_title = [t['title'] for t in jf]
all_times = [t['createdAt'] for t in jf]
all_comments = [c['commentCount'] for c in jf]
all_likes = [l['likeCount'] for l in jf]

for i in range(len(all_title)):
    print(f'Title:{all_title[i]}\nTime:{all_times[i]}\nComments:{all_comments[i]}\nLikes:{all_likes[i]}\n')
```

```python
# 3. è¨ˆç®—ç†±é–€/éžç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
cold_likes= [l['likeCount'] for l in jf2]
cold_comments = [c['commentCount'] for c in jf2]

avg_hot_likes = int(sum(all_likes) / len(all_likes))
avg_hot_comments = int(sum(all_comments) / len(all_comments))
avg_cold_likes = int(sum(cold_likes) / len(cold_likes))
avg_cold_comments = int(sum(cold_comments) / len(cold_comments))

print(f'ç†±é–€ --> average likes : {avg_hot_likes}, average comments : {avg_hot_comments}\n')
print(f'éžç†±é–€ --> average likes : {avg_cold_likes}, average comments : {avg_cold_comments} ')
```
