### 作業part 2：爬取每種貨幣對台幣的一個月歷史匯率資料，並畫出趨勢圖

```python
url = 'https://rate.bot.com.tw/xrt/history/'
res = requests.get(url)
soup = BeautifulSoup(res.text)
country = [i['value'] for i in soup.find('select',attrs={'name':'currency'}).find_all('option')]
```

```python
#1month
date = '2019-11'
for c in country:
    url = f'https://rate.bot.com.tw/xrt/quote/{date}/{c}'
    df = pd.read_html(url)[0]
    df = df.iloc[:,:6]
    df.columns = ['date','currency','cash_buy','cash_sell','spot_buy','spot_sell']
    df.set_index('date',inplace=True)
    df.sort_index(ascending=True,inplace=True)

    #plot
    plt.figure(figsize=(15, 8))
    df[['cash_buy','cash_sell']].plot()  
    plt.legend(loc="best")
    plt.title(c)
    plt.show()
```
