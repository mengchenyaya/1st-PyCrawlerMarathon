## 作業目標

根據範例 ，完成以下問題：

* ① 印出最新文章的「作者」「標題」「時間」
* ② 印出第一頁所有文章的「作者」「標題」「時間」


```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)
soup = BeautifulSoup(r.text)
```

### ① 印出最新文章的「作者」「標題」「時間」

```python
articles = soup.find_all('div',class_='r-ent')
new = articles[-6] #第一頁倒數5個為公告
title = new.find('div',class_='title').text.replace('\n','').replace('\t','')
print(f"!!!NEW!!!\nauthor : {new.find('div',class_='author').text}\ntitle : {title}\ndate : {new.find('div',class_='date').text}")
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
for art in articles[:-6]:
    print(art.find('div',class_='author').text)
    print(art.find('div',class_='title').text.replace('\n','').replace('\t',''))
    print(art.find('div',class_='date').text+'\n')
```

### ③ 試著爬爬看其他版的文章
